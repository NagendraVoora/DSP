```python
country=["germany","india","UK","france"]
gdp=[34300,54400,43400,22300]
plt.bar(country,gdp,color="yellow")
plt.xlabel("country")
plt.ylabel("gdp growth")
plt.title("GDP plot")
plt.show()
```

```python
country=["germany","india","UK","france"]
gdp=[34300,54400,43400,22300]
plt.barh(country,gdp,color="red")
plt.xlabel("country")
plt.ylabel("gdp growth")
plt.title("GDP plot")
plt.show()
```

```python
country=["germany","india","UK","france"]
gdp=[34300,54400,43400,22300]
plt.bar(country,gdp,color=["yellow","red","blue","pink"])
plt.xlabel("country")
plt.ylabel("gdp growth")
plt.title("GDP plot")
plt.show()
```

```python
country=["germany","india","UK","france"]
gdp=[34300,54400,43400,22300]
plt.bar(country,gdp,color=["yellow","red","blue","pink"],ecolor=["black","green","yellow","blue"])
plt.xlabel("country")
plt.ylabel("gdp growth")
plt.title("GDP plot")
plt.show()
```
